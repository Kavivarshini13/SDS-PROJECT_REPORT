## Line Plot
**A line plot** is a type of graph used to display information as a series of data points connected by straight line segments. It is commonly used to visualize trends or changes over time or to show the relationship between two numerical variables

```python
df_cleaned = df.dropna(subset=['display', 'price'])

# Create the line plot for 'Display' vs 'Price'
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='display',  # Display as the x-axis
    y='price',  # Price as the y-axis
    data=df_cleaned,
    marker='o',  # Add markers to each data point
    color='red'  # Line color
)

plt.title('Line Plot: Price vs Display')
plt.xlabel('display Size (inches)')
plt.ylabel('price (INR)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Multivariate Analysis
It is an extension of bivariate analysis which means it involves multiple variables at the same timeto find correlation between them. Multivariate Analysis is a set of statistical model that examine patterns in multidimensional data by considering at once, several data variable.

**PCA** :PCA is a dimensionality reduction technique used in multivariate analysis. It reduces the number of variables while keeping the most important information.


```python
from sklearn import datasets, decomposition 
iris = datasets.load_iris() 
X = iris.data 
y = iris.target 
pca = decomposition.PCA(n_components=2) 
X = pca.fit_transform(X) 
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

## Heat Map
**A heatmap** is a data visualization technique that uses a color-coded matrix to represent the magnitude of values in a dataset. It is commonly used to show relationships, patterns, or correlations between variables.

```python

df_corr = df.select_dtypes(include='number').corr()
sns.heatmap(df_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
